## Hotel Booking Status

```python
# Creating a proper format of date
data['arrival_date'] = data['arrival_date_day_of_month'].astype('str') + '-' + data['arrival_date_month'].astype('str') + '-' + data['arrival_date_year'].astype('str')
```

```python
# Conveting the string to date 
data['arrival_date'] = pd.to_datetime(data['arrival_date'], dayfirst= True)
```

```python
# Extracting Month from date
data['arrival_date_month'] = data['arrival_date'].dt.month
```

```python
# Extracting data for cancelled and confirmed booking by month
booking_status = pd.crosstab(data.arrival_date_month, data.is_canceled)
```

```python
booking_status
```

```python
# Gragh - Booking Status VS months
booking_status.plot()

plt.title('Booking Status vs. Months')
plt.xlabel('Month')
plt.ylabel('Number of bookings')
```

Booking Status Vs. Months graph is the the analysis of the number of confirmed booking and the cancelled booking of each month in the data.

The graph directly indicates the monthly business of the hotel and its cancellation. It can be clearly be observed that there is and spike in the number obooking in the months of August and hence the highest number of cancellation also for under the same month. Minimum number of cancellation can be observed during the start and the end of the year.

## Frequency of visitors over a year

```python
# Extating the data for confirmed booking
confirmed_bookings = data[data['is_canceled'] == 0]
```

```python
confirmed_bookings.shape
```

```python
# Extracting the with confirmed bookings only
confirmed_bookings_only = confirmed_bookings.groupby('arrival_date_month').agg({'arrival_date_month':'count'})
```

```python
confirmed_bookings_only
```

```python
# Ploting a gragh with numbers of booking based on the data
confirmed_bookings_only.plot(kind='bar')

plt.title('Confirmed Booking vs. Months')
plt.xlabel('Month')
plt.ylabel('Number of booking')
```

Confirmed Booking vs. Months graph clearly states that the during the middle of the year the booking ratio is high in comparison to the booking during the start and the end of the year. this can be due to Holiday season which is observed in the most of the countries around the globe.


```python
plt.figure(figsize = (14, 6))

sns.countplot(x = 'arrival_date_month', 
            data = confirmed_bookings, 
            hue = 'hotel', 
            )

plt.title('Confirmed Booking vs. Month - On the Basis of Hotel' ) 
plt.xlabel('Months')
plt.ylabel('Number of bookings')
```

"Confirmed Booking vs. Month - On the Basis of Hotel" gragh gives a similar type of observation as the last gragh. But there the focus is more on the number of confirmed bookings but based on the type of hotel. 

Also, it can be clearly be obsevered that the number of booking is a a bell shaped data, that is the number of booking is low at the start of the event, which gradually increases till the mid of the year and gradually decrease towards the end of the year. On the other hand, the booking for the Resort hotel is almost constant throughtout the year except for the month of July and August.

This gragh gives us and understanding of the busiest months of the year. Hence, the hotel can work towards maintaining there staff bandwidth.

## ADR 

```python
# Data indicating the Average ADR per month
confirmed_bookings.groupby('arrival_date_month').agg({'adr':'mean'})
```

```python
plt.figure(figsize = (14, 6))

sns.lineplot(x = 'arrival_date_month', 
            y = 'adr',
            hue = 'hotel', 
            data = confirmed_bookings, 
            )
plt.title('Monthly ADR')
plt.xlabel('Months')

plt.tight_layout()
```

The graph displaced the Average Daily Rate(ADR) of the hotel. ADR is the measure of the average paid for rooms sold in a given period. For City Hotel, a gradual increase can be observed during the middle of the year. For Resort Hotel, a sudden spike can be observed from the mid of the year, which is highest for the month of August.
